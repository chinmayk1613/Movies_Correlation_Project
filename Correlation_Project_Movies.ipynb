```python
# to generate scatter plot for gross with budget


plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Relation')
plt.xlabel('Gross')
plt.ylabel('Budget')
plt.show()

```

```python
#Analyze the correlation #pearson, kendall, spearman

df.corr(method='pearson')
```

```python
# we can conclude from above table there is high correlation between budget and gross


corr_matrix=df.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrics Numeric data')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()


```

```python
# To make numerize the comapany name

df_num=df

for column_name in df_num.columns:
    if(df_num[column_name].dtype == 'object'):
        df_num[column_name]=df_num[column_name].astype('category')
        df_num[column_name]=df_num[column_name].cat.codes
df_num.head()
```

```python
df_num.corr()
```

```python
cor_mat=df_num.corr()

corr_pairs=cor_mat.unstack()

corr_pairs
```

```python
corr_pairs_sorted=corr_pairs.sort_values()

corr_pairs_sorted
```

```python
high_correlation=corr_pairs_sorted[(corr_pairs_sorted) > 0.5]

high_correlation
```
